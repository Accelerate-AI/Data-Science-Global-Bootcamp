We can try the following two encoding techniques.
- Label Encoding
- One Hot Encoding

Here "Region" names do not have an order or rank. When Label Encoding is performed, the region names are ranked based on alphabets. Due to this, there is a very high probability that the model captures the relationship between regions such as "America" < "East Asia & Pacific" < "Sub-Saharan Africa" OR may be something like an unknown/unjustified sequence. This is something we do not want.

We can overcome this with the help of "One Hot Encoding" method.


#### Using Label Encoding

```python
df_factorize = df_assignment.copy()

df_factorize.sample(3)
```

```python
df_factorize.loc[:,'Region_FactorizeEnc'] = pd.factorize(df_factorize['Region'])[0].reshape(-1,1)

df_factorize.head(6)
```

#### Using One Hot Encoding

```python
df_assignment_ohe = pd.get_dummies(df_assignment, columns = ['Region'])

df_assignment_ohe
```

Let's refer free “Wine” Dataset that is deposited on the UCI machine learning repository. You can refer to GitHub for this - https://github.com/Accelerate-AI/Data-Science-Global-Bootcamp/blob/main/01%20Python/wine_data_UCI.csv  The Wine dataset consists of 3 different classes/qualities where each row corresponds to a particular wine sample. 

The "quality" features indicate the class/quality of wine and it is represented as (1, 2, 3) and rest of the columns correspond to 13 different attributes (features).

Consider the features - the wine quality i.e. "quality", "Alcohol" (percent/volume) and "Malicacid" (g/l).

a) Do you think feature scaling is required? If yes - Why, If no - why?

b) If you feel feature scaling is required, then perform Standardization and Normalization and provide your result. What is the difference you observe between these two methods? 

### Solution:

```python
import pandas as pd
import numpy as np

from sklearn import preprocessing
```

```python
df = pd.read_csv("./wine_data_UCI.csv")

df.sample(5)
```

```python
df.shape
```

```python
df1=df.filter(['quality', 'Alcohol', 'Malicacid'])

df1.sample(5)
```

As we can see in the table above, the features Alcohol (percent/volumne) and Malic acid (g/l) are measured on different scales, so that Feature Scaling is necessary important prior to any comparison or combination of these data.

```python
std_scale = preprocessing.StandardScaler().fit(df1[['Alcohol', 'Malicacid']])
df_std = std_scale.transform(df1[['Alcohol', 'Malicacid']])
```

```python
df_std
```

```python
minmax_scale = preprocessing.MinMaxScaler().fit(df1[['Alcohol', 'Malicacid']])
df_minmax = minmax_scale.transform(df1[['Alcohol', 'Malicacid']])
```

```python
df_minmax
```

We can clearly see that when Min-Max scaling is performed the values are normalized between a range from 0 to 1. It is different in the standard scaler case where standrdization is performed.
