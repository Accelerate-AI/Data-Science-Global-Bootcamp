#### Tree Based Models

In this notebook we will look at building decision trees using 2 method:
 - ID3
 - CART

```python
import numpy as np
import pandas as pd

from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("play.csv")
```

```python
df.head()
```

```python
X = pd.get_dummies(df.drop("Decision", axis=1), drop_first=True)
y = df["Decision"]
```

```python
y.value_counts()
```

```python
# CART Model for classification
model = DecisionTreeClassifier(criterion='gini')
model.fit(X, y)

plt.figure(figsize=(6,6), dpi=150)
plot_tree(model, feature_names=X.columns, filled=True);
```

```python
plt.figure(figsize=(6, 4))
plt.bar(np.arange(X.shape[1]), np.squeeze(np.abs(model.feature_importances_)) )
plt.xticks(np.arange(X.shape[1]), ['Outlook_Rain','Outlook_Sunny', 'Temp_Hot',
                                  'Temp_Mild',  'Humidity_Normal', 'Wind_Weak'],
                     rotation = 90)
None
```

```python
#ID3 Model 
id3_model = DecisionTreeClassifier(criterion='entropy')
id3_model.fit(X, y)

plt.figure(figsize=(6,6), dpi=150)
plot_tree(id3_model, feature_names=X.columns, filled=True);
```

```python
plt.figure(figsize=(6, 4))
plt.bar(np.arange(X.shape[1]), np.squeeze(np.abs(id3_model.feature_importances_)) )
plt.xticks(np.arange(X.shape[1]), ['Outlook_Rain','Outlook_Sunny', 'Temp_Hot',
                                  'Temp_Mild',  'Humidity_Normal', 'Wind_Weak'],
                     rotation = 90)
None
```
